```python
# This program is into GOOGLE COLAB due to some GPU and Libray Loading issues in Local machine
# --------------------------------------------------------------------------------------------
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn import metrics

# Saperatly againg generating 5000 synthetic data points for this istance of code
# because it is on GOOGLE COLAB
vector_x = random.normal(loc=0, scale=1, size=5000)
vector_esp = random.normal(loc=0, scale=0.25, size=5000)
vector_y = -1 + 0.5 * vector_x - 2 * (vector_x ** 2) + 0.3 * (vector_x ** 3) + vector_esp
```

```python
# reshaping dataset to fit into model 
X = vector_x.reshape(-1, 1)
Y = vector_y.reshape(-1, 1)

# initiating model with help of TensorFlow Keras
model = tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=(1,), activation='tanh')])
model.compile(optimizer=tf.keras.optimizers.SGD(0.01), loss='mean_squared_error', metrics=['mean_squared_error'])

# fitting/traning model on synthetic data points
model.fit(X, Y, epochs=30, batch_size=512)
prediction = model.predict(X)

# printing numeric data of our sigmod function
print(" \n Sigmoid Nueron using tanh activation function: \n ")
print("----------------------------------------------------------")
mean_sq_error = metrics.mean_squared_error(vector_y, prediction)
print(" \n Result:")
print("------------")
print("Mean Square Error: ", mean_sq_error)
new_data = np.linspace(vector_x.min(), vector_x.max())

# plotting graph of our synthetic data points used in sigmod function
plt.scatter(X, vector_y, color='green', label='Data Points', marker='.')
plt.plot(new_data, model.predict(new_data.reshape(-1)), color='blue', label='Regression Line')
plt.title("Sigmoid Nueron - tanh Activation")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
